```python
from kaggle_environments import make
import numpy as np
import math
import torch
import torch.nn as nn
import torch.optim as optim
from diffusers import DDPMScheduler
from tqdm.notebook import tqdm
import random

# Xác định tên môi trường chính xác
ENV_NAME = "orbit_wars"
env = make(ENV_NAME)
states = env.reset()
obs = states[0]['observation']
print("Môi trường OK. Observation keys:", obs.keys())
print("Player:", obs['player'])
```

```python
MACRO_ACTIONS = ['expand', 'aggressive', 'dominate', 'counter_attack',
                 'rush', 'grow', 'smash', 'defend']

phase_to_macro = {
    'expand': 0, 'aggressive': 1, 'dominate': 2, 'counter_attack': 3,
    'rush': 4, 'grow': 5, 'smash': 6, 'defend': 7, 'crush': 1,
    'opportunistic': 0, 'wait': 7, '': 7
}

def phase_to_index(phase):
    return phase_to_macro.get(phase, 7)

def extract_state(obs):
    player = obs['player']
    planets = obs['planets']
    fleets = obs.get('fleets', [])
    step = obs.get('step', 0)
    omega = obs.get('angular_velocity', 0.03)

    my = [p for p in planets if p[1] == player]
    enemy = [p for p in planets if p[1] not in (player, -1)]
    neutral = [p for p in planets if p[1] == -1]

    my_ships = sum(p[5] for p in my)
    enemy_ships = sum(p[5] for p in enemy)

    my_prod = sum(p[6] for p in my)
    enemy_prod = sum(p[6] for p in enemy)

    # ratios (log-stable)
    ship_ratio = np.log1p(my_ships) - np.log1p(enemy_ships + 1e-8)
    prod_ratio = np.log1p(my_prod) - np.log1p(enemy_prod + 1e-8)

    # fleet pressure
    my_fleets = sum(f[6] for f in fleets if f[1] == player)
    enemy_fleets = sum(f[6] for f in fleets if f[1] != player and f[1] != -1)

    fleet_pressure = enemy_fleets - my_fleets

    # spatial signals
    def min_dist(A, B):
        if not A or not B: return 0
        return min(math.hypot(a[2]-b[2], a[3]-b[3]) for a in A for b in B)

    min_enemy_dist = min_dist(my, enemy)
    min_neutral_dist = min_dist(my, neutral)

    # threat density
    enemy_nearby = 0
    for p in my:
        if any(math.hypot(p[2]-e[2], p[3]-e[3]) < 30 for e in enemy):
            enemy_nearby += 1
    enemy_nearby /= (len(my) + 1e-8)

    # distribution
    ship_std = np.std([p[5] for p in planets])
    prod_std = np.std([p[6] for p in planets])

    # time encoding
    sin_t = np.sin(step * 0.01)
    cos_t = np.cos(step * 0.01)

    features = [
        len(my)/20, len(enemy)/20, len(neutral)/20,
        my_ships/500, enemy_ships/500,
        my_prod/30, enemy_prod/30,
        ship_ratio, prod_ratio,
        fleet_pressure/300,
        min_enemy_dist/100,
        enemy_nearby,
        ship_std/50,
        prod_std/10,
        omega,
        sin_t, cos_t
    ]

    features = np.array(features, dtype=np.float32)
    return np.pad(features, (0, 32-len(features)))

print(" State extractor OK")
```

```python
def random_agent(obs):
    """Trả về một hành động ngẫu nhiên hợp lệ."""
    planets = obs['planets']
    player = obs['player']
    my_planets = [p for p in planets if p[1] == player]
    if not my_planets:
        return []
    moves = []
    for src in my_planets[:3]:  # giới hạn
        if src[5] > 10:
            angle = random.uniform(0, 2*math.pi)
            ships = random.randint(5, min(50, src[5]))
            moves.append([src[0], angle, ships])
    return moves
```

```python
def collect_dataset(num_episodes=500, max_steps=500):
    dataset = []
    for ep in tqdm(range(num_episodes), desc="Collecting"):
        env = make(ENV_NAME)
        states = env.reset()
        obs_me = states[0]['observation']
        done = False
        steps = 0
        while not done and steps < max_steps:
            moves, phase = agent(obs_me)
            state = extract_state(obs_me)
            macro = phase_to_index(phase)
            dataset.append((state, macro))

            # Đối thủ random
            obs_opp = states[1]['observation']
            opp_action = random_agent(obs_opp)

            next_states = env.step([moves, opp_action])
            next_state_me = next_states[0]
            obs_me = next_state_me['observation']
            done = (next_state_me['status'] != 'ACTIVE')
            states = next_states
            steps += 1
        if (ep+1) % 100 == 0:
            print(f"Episode {ep+1}: total samples={len(dataset)}")
    return dataset

print("Bắt đầu thu thập...")
dataset = collect_dataset(2000)
states = np.array([d[0] for d in dataset])
actions = np.array([d[1] for d in dataset])
np.savez("offline_dataset.npz", states=states, actions=actions)
print(f"Đã lưu {len(dataset)} mẫu")
```

```python
# Cell 7: Huấn luyện Diffusion Policy (đã sửa lỗi device)
data = np.load("combined_dataset.npz")
states = data["states"]
actions = data["actions"]
num_classes = 8

actions_onehot = np.zeros((actions.shape[0], num_classes), dtype=np.float32)
actions_onehot[np.arange(actions.shape[0]), actions] = 1.0

split = int(0.9 * len(states))
train_states = torch.tensor(states[:split], dtype=torch.float32)
train_actions = torch.tensor(actions_onehot[:split], dtype=torch.float32)
val_states = torch.tensor(states[split:], dtype=torch.float32)
val_actions = torch.tensor(actions_onehot[split:], dtype=torch.float32)
print(f"Train: {len(train_states)}, Val: {len(val_states)}")

class DiffusionPolicy(nn.Module):
    def __init__(self, state_dim=32, action_dim=8, hidden_dim=256):
        super().__init__()
        self.time_mlp = nn.Sequential(
            nn.Linear(1, hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim),
        )
        self.net = nn.Sequential(
            nn.Linear(state_dim + action_dim + hidden_dim, hidden_dim),
            nn.ReLU(),
            nn.Linear(hidden_dim, hidden_dim), nn.ReLU(),
            nn.Linear(hidden_dim, action_dim),
        )
    def forward(self, state, action_noisy, timestep):
        t_emb = self.time_mlp(timestep.unsqueeze(-1).float())
        x = torch.cat([state, action_noisy, t_emb], dim=-1)
        return self.net(x)

scheduler = DDPMScheduler(num_train_timesteps=100, beta_schedule="squaredcos_cap_v2")
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = DiffusionPolicy().to(device)
optimizer = optim.Adam(model.parameters(), lr=1e-3)

# Đưa dữ liệu validation lên device ngay từ đầu
val_states = val_states.to(device)
val_actions = val_actions.to(device)

batch_size = 256
epochs = 250
for epoch in range(epochs):
    model.train()
    total_loss = 0
    perm = torch.randperm(len(train_states))
    for i in range(0, len(train_states), batch_size):
        idx = perm[i:i+batch_size]
        s = train_states[idx].to(device)
        a = train_actions[idx].to(device)
        noise = torch.randn_like(a)
        t = torch.randint(0, scheduler.config.num_train_timesteps, (len(idx),), device=device)
        a_noisy = scheduler.add_noise(a, noise, t)
        noise_pred = model(s, a_noisy, t.float() / scheduler.config.num_train_timesteps)
        loss = nn.functional.mse_loss(noise_pred, noise)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / max(1, len(train_states)//batch_size)

    # Validation
    model.eval()
    with torch.no_grad():
        val_t = torch.randint(0, scheduler.config.num_train_timesteps, (len(val_states),), device=device)
        val_noise = torch.randn_like(val_actions)  # lúc này val_actions đã ở device
        val_a_noisy = scheduler.add_noise(val_actions, val_noise, val_t)
        val_pred = model(val_states, val_a_noisy, val_t.float() / scheduler.config.num_train_timesteps)
        val_loss = nn.functional.mse_loss(val_pred, val_noise)
    print(f"Epoch {epoch+1}: loss={avg_loss:.4f}, val_loss={val_loss.item():.4f}")

torch.save(model.state_dict(), "diffusion_policy.pth")
print("Model saved")
```
